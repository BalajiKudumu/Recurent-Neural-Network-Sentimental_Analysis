```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentence=['the glass of milk',
          'the glass of juice',
          'the cup of tea',
          'I am a good boy',
          'I am a good developer',
          'understand the meaning of words',
          'My posters are good',
]
```

```python
# define the vocabulary size
voc_size=10000
```

```python
## One hot representation
onehot_representation=[one_hot(word,voc_size )for word in sentence]
onehot_representation
```

```python
## Word Embedding representation
import numpy as np
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
sent_length=8
embedded_doc=pad_sequences(onehot_representation,padding='pre',maxlen=sent_length)
print(embedded_doc)
```

```python
## Feature Representation
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim))
model.build(input_shape=(None, sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_doc)
```
